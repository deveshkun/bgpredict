```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from collections import Counter
import numpy as np
import os
import shutil
from sklearn.utils import resample
from tensorflow.keras.preprocessing.image import load_img,img_to_array,save_img
from tensorflow.keras.utils import image_dataset_from_directory
from sklearn.metrics import classification_report,confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns


dataset_path = r"D:\PythonProject\dataset_blood_group"
batch_size = 32


# Load Dataset

dataset = image_dataset_from_directory(
    dataset_path,
    labels = "inferred",
    label_mode = "int",
    image_size = (64, 64),
    batch_size = batch_size,
    shuffle = True,
)

# Class Distribution

class_name = dataset.class_names
class_count = Counter()

for _,labels in dataset.unbatch():
    class_count[int(labels.numpy())] += 1

print("Class Distributions")
for i ,count in class_count.items():
    print(f"{class_name[i]}: {count}")

def plot_class_distribution(class_name, class_count):
    classes = [class_name[i] for i in class_count.keys()]
    counts = [class_count[i] for i in class_count.keys()]

    plt.figure(figsize = (10,6))
    plt.bar(classes, counts)
plot_class_distribution(class_name, class_count)


# Oversampling
max_count = max(class_count.values())
def oversample(class_id,count,max_count):

    unbatched_dataset = dataset.unbatch()

    class_dataset = unbatched_dataset.filter(lambda img,lbl: tf.equal(lbl,class_id))

    repeat_factor = max_count // count+(max_count % count > 0)

    return class_dataset.repeat(repeat_factor).take(max_count)

balanced_datasets = []

# Oversample each class and store datasets
for class_id, count in class_count.items():
    oversampled_ds = oversample(class_id, count, max_count)  # Ensure this returns a tf.data.Dataset
    balanced_datasets.append(oversampled_ds)

# Combine the oversampled datasets
balanced_dataset = tf.data.Dataset.sample_from_datasets(balanced_datasets)

# Count class distribution
balance_class_count = Counter(int(lbl.numpy()) for _, lbl in balanced_dataset)

# Plot class distribution
plot_class_distribution(class_name, balance_class_count)

# Apply batching
balanced_dataset = balanced_dataset.batch(batch_size, drop_remainder=True)


for sample in balanced_dataset.take(10):
    print(sample[0].shape)

balanced_dataset_unbatched = balanced_dataset.unbatch()
dataset_size = sum(1 for _ in balanced_dataset_unbatched)
print(f"Dataset Size: {dataset_size}")

balanced_dataset_unbatched = balanced_dataset.unbatch()

train_ratio,val_ratio = 0.7,0.2

train_size = int(train_ratio *dataset_size)
val_size = int(val_ratio *dataset_size)

train_dataset = balanced_dataset_unbatched.take(train_size)
val_test_dataset = balanced_dataset_unbatched.skip(train_size)

val_dataset = val_test_dataset.take(val_size)
test_dataset = val_test_dataset.skip(val_size)

train_dataset = train_dataset.batch(batch_size, drop_remainder=True)
val_dataset = val_dataset.batch(batch_size, drop_remainder=True)
test_dataset = test_dataset.batch(batch_size,drop_remainder=True)

train_batch_Count = sum(1 for _ in train_dataset)
val_batch_Count = sum(1 for _ in val_dataset)
test_batch_Count = sum(1 for _ in test_dataset)

print(f"Train Batch Count: {train_batch_Count * batch_size}")
print(f"Val Batch Count: {val_batch_Count * batch_size}")
print(f"Test Batch Count: {test_batch_Count * batch_size}")

```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Dropout, Flatten, Dense

def hybrid_model():
    input_layer = Input(shape=(64, 64, 3))

    # CNN Feature Extractor
    x = Conv2D(32, (3,3), activation="relu", padding='same')(input_layer)
    x = MaxPooling2D(2,2)(x)
    x = Dropout(0.3)(x)

    x = Conv2D(64, (3,3), activation="relu", padding='same')(x)
    x = MaxPooling2D(2,2)(x)
    x = Dropout(0.4)(x)

    x = Conv2D(128, (3,3), activation="relu", padding='same')(x)
    x = MaxPooling2D(2,2)(x)
    x = Dropout(0.4)(x)

    x = Conv2D(256, (3,3), activation="relu", padding='same')(x)
    x = MaxPooling2D(2,2)(x)
    x = Dropout(0.4)(x)

    x = Conv2D(512, (3,3), activation="relu", padding='same')(x)
    x = MaxPooling2D(2,2)(x)
    x = Dropout(0.5)(x)

    x = Flatten()(x)

    # ANN Classifier
    x = Dense(512, activation="relu")(x)
    x = Dropout(0.5)(x)
    x = Dense(256, activation="relu")(x)
    x = Dropout(0.4)(x)
    output_layer = Dense(len(class_name), activation="softmax")(x)

    model = Model(inputs=input_layer, outputs=output_layer)

    model.compile(optimizer="adam",
                  loss="sparse_categorical_crossentropy",
                  metrics=["accuracy"])
    return model

create_model = hybrid_model()

from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
reduce_lr = ReduceLROnPlateau(monitor="val_loss",
                              factor=0.5,
                              patience=3,
                              verbose=1,
                              min_lr=1e-6)

early_stopping = EarlyStopping(patience=5,
                               monitor="val_loss",
                               verbose=1,
                               restore_best_weights=True)

# train

history = create_model.fit(train_dataset,
                           validation_data=val_dataset,
                           epochs=50,
                           callbacks=[reduce_lr,early_stopping])


create_model.save("model2.keras")
print("Model saved")
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Prediction
y_true = []
y_pred = []

for images, labels in test_dataset:
    predictions = create_model.predict(images)
    predict_labels = np.argmax(predictions, axis=1)
    y_true.extend(labels.numpy())
    y_pred.extend(predict_labels)

# Evaluation
y_true = np.array(y_true)
y_pred = np.array(y_pred)

report = classification_report(y_true, y_pred, target_names=class_name)
print(report)

# ❗️Use a new variable name instead of overwriting the function
cm = confusion_matrix(y_true, y_pred)

# Plot confusion matrix
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_name, yticklabels=class_name)
plt.title('Confusion Matrix')
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```

```python
def plot_accuracy(history):
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'val'], loc='upper left')
    plt.show()

plot_accuracy(history)
```
